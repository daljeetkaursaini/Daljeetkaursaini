```python
import cv2
import numpy as np

```

```python
image = cv2.imread("th.jpg")
```

```python



image.shape

corner=image[0:300,0:300]

cv2_imshow(corner)


canvas = np.zeros((300, 300, 3), dtype = "uint8")

```

```python
M = np.float32([[1, 0, 25], [0, 1, 50]])
print(image.shape[0])
print(image.shape[1])
shifted = cv2.warpAffine(image, M, (image.shape[1]+30, image.shape[0]+55))
cv2_imshow(shifted)
cv2.waitKey(0)

```

```python
cv2_imshow(image)
(h, w) = image.shape[:2]
```

```python
green = (0, 255, 0)
cv2.line(canvas, (0, 0), (300, 300), green,4)
cv2_imshow(canvas)
cv2.waitKey(0)

red = (0, 0, 255)
blue = (255, 0, 0)
cv2.line(canvas, (300, 0), (0, 300), red, 3)   
cv2.rectangle(canvas, (300, 0), (0, 300), blue, 3)
cv2.circle(canvas, (150,150), 120, blue, 4)
cv2_imshow(canvas)
cv2.waitKey(0)
```

```python

cv2_imshow(image)
(h, w) = image.shape[:2]


center = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(center, 45, 1.0)
print(M)
rotated = cv2.warpAffine(image, M, (w, h))
cv2_imshow(rotated)
cv2.waitKey(0)

cv2_imshow(image)
print(image.shape)
r = 150.0 / image.shape[1]
dim = (150, int(image.shape[0] * r))
resized = cv2.resize(image, dim, interpolation = cv2.INTER_AREA)
print(resized.shape)
cv2_imshow(resized)
cv2.waitKey(0)

cv2_imshow(image)
flipped = cv2.flip(image, 0)
flipped1 = cv2.flip(image, 1)
cv2_imshow(flipped)
cv2_imshow(flipped1)
cv2.waitKey(0)

```
